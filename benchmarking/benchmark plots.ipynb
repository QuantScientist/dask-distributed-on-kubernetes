```python
import pandas, seaborn, matplotlib
from matplotlib import pyplot
%matplotlib inline
```

```python
df = pandas.read_csv("../kubernetes/results_benchmark3.raw", header=None)
df.columns = ["indicator", "workers", "replica", "tasks", "input", "task_time", "allocate", "output", "runtime"]
del df["indicator"]

seaborn.set_context('talk')
seaborn.pointplot(data=df, x='workers', y='runtime')
pyplot.title("Time to run %d tasks, each task takes %g sec" % (df.tasks.unique()[0], df.task_time.unique()[0] ))
pyplot.xlabel("Num workers")
pyplot.ylabel("Total time (sec)")
pyplot.ylim(ymin=0)

df
```
